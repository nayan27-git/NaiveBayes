```python
df_train = df.iloc[:5333]
```

```python
p_yes = df_train[df_train['Play']=='Yes'].shape[0]/df_train.shape[0]
#K=number of unique x in p(x|y) alpha=1

p_yes_Outlook_overcast = (df_train[(df_train['Play']=='Yes') & (df_train['Outlook']=='Overcast')].shape[0]+1)/(df_train[df_train['Play']=='Yes'].shape[0]+3)
p_yes_Outlook_sunny = (df_train[(df_train['Play']=='Yes') & (df_train['Outlook']=='Sunny')].shape[0]+1)/(df_train[df_train['Play']=='Yes'].shape[0]+3)
p_yes_Outlook_rainy = (df_train[(df_train['Play']=='Yes') & (df_train['Outlook']=='Rainy')].shape[0]+1)/(df_train[df_train['Play']=='Yes'].shape[0]+3)
p_yes_Temperature_cool = (df_train[(df_train['Play']=='Yes') & (df_train['Temperature']=='Cool')].shape[0]+1)/(df_train[df_train['Play']=='Yes'].shape[0]+3)
p_yes_Temperature_mild = (df_train[(df_train['Play']=='Yes') & (df_train['Temperature']=='Mild')].shape[0]+1)/(df_train[df_train['Play']=='Yes'].shape[0]+3)
p_yes_Temperature_hot = (df_train[(df_train['Play']=='Yes') & (df_train['Temperature']=='Hot')].shape[0]+1)/(df_train[df_train['Play']=='Yes'].shape[0]+3)
p_yes_Humidity_normal = (df_train[(df_train['Play']=='Yes') & (df_train['Humidity']=='Normal')].shape[0]+1)/(df_train[df_train['Play']=='Yes'].shape[0]+2)
p_yes_Humidity_high = (df_train[(df_train['Play']=='Yes') & (df_train['Humidity']=='High')].shape[0]+1)/(df_train[df_train['Play']=='Yes'].shape[0]+2)
p_yes_Wind_weak = (df_train[(df_train['Play']=='Yes') & (df_train['Wind']=='Weak')].shape[0]+1)/(df_train[df_train['Play']=='Yes'].shape[0]+2)
p_yes_Wind_strong = (df_train[(df_train['Play']=='Yes') & (df_train['Wind']=='Strong')].shape[0]+1)/(df_train[df_train['Play']=='Yes'].shape[0]+2)
```

```python
p_No = df_train[df_train['Play']=='No'].shape[0]/df_train.shape[0]

p_No_Outlook_overcast = (df_train[(df_train['Play']=='No') & (df_train['Outlook']=='Overcast')].shape[0]+1)/(df_train[df_train['Play']=='No'].shape[0]+3)
p_No_Outlook_sunny = (df_train[(df_train['Play']=='No') & (df_train['Outlook']=='Sunny')].shape[0]+1)/(df_train[df_train['Play']=='No'].shape[0]+3)
p_No_Outlook_rainy = (df_train[(df_train['Play']=='No') & (df_train['Outlook']=='Rainy')].shape[0]+1)/(df_train[df_train['Play']=='No'].shape[0]+3)
p_No_Temperature_cool = (df_train[(df_train['Play']=='No') & (df_train['Temperature']=='Cool')].shape[0]+1)/(df_train[df_train['Play']=='No'].shape[0]+3)
p_No_Temperature_mild = (df_train[(df_train['Play']=='No') & (df_train['Temperature']=='Mild')].shape[0]+1)/(df_train[df_train['Play']=='No'].shape[0]+3)
p_No_Temperature_hot = (df_train[(df_train['Play']=='No') & (df_train['Temperature']=='Hot')].shape[0]+1)/(df_train[df_train['Play']=='No'].shape[0]+3)
p_No_Humidity_normal = (df_train[(df_train['Play']=='No') & (df_train['Humidity']=='Normal')].shape[0]+1)/(df_train[df_train['Play']=='No'].shape[0]+2)
p_No_Humidity_high = (df_train[(df_train['Play']=='No') & (df_train['Humidity']=='High')].shape[0]+1)/(df_train[df_train['Play']=='No'].shape[0]+2)
p_No_Wind_weak = (df_train[(df_train['Play']=='No') & (df_train['Wind']=='Weak')].shape[0]+1)/(df_train[df_train['Play']=='No'].shape[0]+2)
p_No_Wind_strong = (df_train[(df_train['Play']=='No') & (df_train['Wind']=='Strong')].shape[0]+1)/(df_train[df_train['Play']=='No'].shape[0]+2)
```

```python
df_test = df.iloc[5333:]
df_test.head()
```

```python
df_test['Pred'] = [None]*len(df_test)
df_test['Prob_yes'] = [None]*len(df_test)
df_test.head()
```

```python
print(df_test.iloc[1,2])
```

```python
# print(var_dict[df_test.iloc[0,2]][1])
```

```python
var_dict = {
    'Overcast':[p_No_Outlook_overcast,p_yes_Outlook_overcast],
           'Sunny':[p_No_Outlook_sunny,p_yes_Outlook_sunny],
           'Rainy':[p_No_Outlook_rainy,p_yes_Outlook_rainy],
           'Cool':[p_No_Temperature_cool,p_yes_Temperature_cool],
           'Mild':[p_No_Temperature_mild,p_yes_Temperature_mild],
           'Hot':[p_No_Temperature_hot,p_yes_Temperature_hot],
           'Normal':[p_No_Humidity_normal,p_yes_Humidity_normal],
           'High':[p_No_Humidity_high,p_yes_Humidity_high],
           'Weak':[p_No_Wind_weak,p_yes_Wind_weak],
           'Strong':[p_No_Wind_strong,p_yes_Wind_strong]
           }
for i in range(len(df_test)):
    p_y = p_yes
    p_n = p_No
    for j in range(1,5):  
        p_y *= var_dict[df_test.iloc[i,j]][1]
        p_n *= var_dict[df_test.iloc[i,j]][0]
    Normalised_p_y = p_y/(p_y+p_n)
    Normalised_p_n = 1-Normalised_p_y
    df_test.iloc[i,6] = ('Yes' if Normalised_p_y > Normalised_p_n else 'No')
    df_test.iloc[i, df_test.columns.get_loc('Prob_yes')] = Normalised_p_y 
```

```python
df_test.head(50)
```

```python
mapping = {'Yes':1,'No':0}
df_test['play_enc'] = df_test['Play'].map(mapping)
df_test['pred_enc'] = df_test['Pred'].map(mapping)
df_test.head()
```

```python
print(len(df_test))
```

```python
p_predicted = df_test['Prob_yes'].astype(float) 

# 3. Recalculate the average cost (Log Loss)
# cost_fn = -ylog(y_pred)-(1-y)log(1-y_pred)

avg_cost_fn = np.sum(-df_test['play_enc'] * np.log(p_predicted) - (1 - df_test['play_enc']) * np.log(1 - p_predicted)) / len(df_test)

print(avg_cost_fn)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
accuracy = accuracy_score(df_test['play_enc'],df_test['pred_enc'])
precision = precision_score(df_test['play_enc'],df_test['pred_enc'])
recall = recall_score(df_test['play_enc'],df_test['pred_enc'])
cm = confusion_matrix(df_test['play_enc'],df_test['pred_enc'])
print(cm)
print(accuracy,precision,recall)                          
```

```python
df_train = df.iloc[:5333]
df_test = df.iloc[5333:]
X_train = df_train.drop(['Day', 'Play'], axis=1)
y_train = df_train['Play']
X_test = df_test.drop(['Day', 'Play'], axis=1)
y_test = df_test['Play']
```

```python
outlook_mode = X_train['Outlook'].mode()[0]
temp_mode = X_train['Temperature'].mode()[0]
humidity_mode = X_train['Humidity'].mode()[0]
wind_mode = X_train['Wind'].mode()[0]
```

```python
X_train['Outlook'] = X_train['Outlook'].fillna(outlook_mode)
X_train['Temperature'] = X_train['Temperature'].fillna(temp_mode)
X_train['Humidity'] = X_train['Humidity'].fillna(humidity_mode)
X_train['Wind'] = X_train['Wind'].fillna(wind_mode)

X_test['Outlook'] = X_test['Outlook'].fillna(outlook_mode)
X_test['Temperature'] = X_test['Temperature'].fillna(temp_mode)
X_test['Humidity'] = X_test['Humidity'].fillna(humidity_mode)
X_test['Wind'] = X_test['Wind'].fillna(wind_mode)
```

```python
X_train = pd.get_dummies(X_train, columns=['Outlook','Temperature','Humidity','Wind'], drop_first=True, dtype=np.int64)
X_test = pd.get_dummies(X_test, columns=['Outlook','Temperature','Humidity','Wind'], drop_first=True, dtype=np.int64)
```

```python
y_test.shape
```

```python
X_train.head()
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train, y_train)
```

```python
y_nb = nb.predict(X_test)
y_prob = nb.predict_proba(X_test)
```

```python
y_nb
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, log_loss, f1_score
print("Accuracy:",accuracy_score(y_test, y_nb))
print("Precision:",precision_score(y_test, y_nb, pos_label='Yes'))
print("Recall:",recall_score(y_test, y_nb, pos_label='Yes'))
print("F1:",f1_score(y_test, y_nb, pos_label='Yes'))
print("Binary Crossentropy:",log_loss(y_test, y_prob))
```
